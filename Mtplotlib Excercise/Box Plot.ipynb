```python
from matplotlib import pyplot as plt
```

Visualization of Continuous Numerical Data

Data Visualization of Height ,Weight and Temperature

```python
heig = [120,110,156,144,170]
weig = [40,55,60,70,65]
temp = [37,37.5,38,38.5,39]
y = [heig,weig,temp]
z = ['Height(Cm)','weight(Kg)','Body Temperature(degree C)']
plt.boxplot(y,tick_labels=z)
plt.title("Box Plot of Height,Weight and Body Temperature")
plt.ylabel("Values")
plt.show()
```
