```python
from matplotlib import pyplot as plt
#import numpy as np
## OR
import random
```

```python
x = np.random.randint(10,30,15)
x
```

```python
plt.hist(x,label="Python",color='g',edgecolor='black',linewidth=2)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.legend()
plt.show()
```

```python
plt.hist(x,label="Python",color='g',edgecolor='black',linewidth=2,alpha=0.5)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.hist(x,align='mid',rwidth=0.8,orientation='horizontal')
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.hist(x,log=True,histtype='step')
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.hist(x,cumulative=True)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.hist(x,bottom=10)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.hist(x,bins=20)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.axvline(20,color='r')
plt.show()
```

```python
plt.hist(x,range=[1,40],density=True)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.hist(x,stacked=True,cumulative=-1)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
l=[10,15,20,25,30,35,40,45,50,55,60,65,70,75,80]
plt.hist(x,bins=l)
plt.title("Normal Histogram")
plt.xlabel("VALUES")
plt.ylabel("FREQUENCY")
plt.show()
```
