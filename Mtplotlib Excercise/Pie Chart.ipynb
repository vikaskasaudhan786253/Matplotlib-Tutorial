```python
import matplotlib.pyplot as plt
```

```python
labels=['Apples','Bananas','Mangoes','Grapes']
sizes=[20,40,30,10]
```

```python
plt.pie(sizes,labels=labels)
plt.title("Fruit Distribution")
plt.show()
```

```python
ex = [0.1,0.0,0.0,0.0]
plt.pie(sizes,labels=labels,explode=ex,autopct=None)
plt.title("Fruit Distribution")
plt.show()
```

```python
c=['r','y','g','m']
plt.pie(sizes,labels=labels,explode=None,colors=c,autopct='%0.2f%%',shadow=True,radius=1)
plt.title("Fruit Distribution")
plt.show()
```

```python
c=['r','y','g','m']
plt.pie(sizes,labels=labels,explode=None,colors=c,autopct='%30.2f%%',shadow=True,radius=1)
plt.title("Fruit Distribution")
plt.show()
```

```python
plt.pie(sizes,labels=labels,frame=True,labeldistance=1.1)
plt.title("Fruit Distribution")
plt.show()
```

```python
plt.pie(sizes,labels=labels,frame=True,labeldistance=1.1)
plt.title("Fruit Distribution")
plt.axis('equal')
plt.show()
```
