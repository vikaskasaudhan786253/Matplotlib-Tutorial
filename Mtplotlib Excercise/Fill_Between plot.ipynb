```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
```

```python
plt.plot(x,y,color='r')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',where=(x>=2)&(x<=4))
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',where=(x>=2.3)&(x<=4.3))
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',where=(x>=2.3)&(x<=4.3),interpolate=True)
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',step='pre')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',step='post')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',step='mid')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',linewidth=20)
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='/')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='\\')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='-')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='+')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='x')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='o')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='.')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,hatch='*',edgecolor='k',facecolor='w')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',label='Straight line')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',label='Straight line')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Normal Straight Line Plot')
plt.legend()
plt.show()
```
