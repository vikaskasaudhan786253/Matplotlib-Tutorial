```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Colored Scatterplot

```python
iris = pd.read_csv('iris.csv')
iris.sample(10)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
## cmap and alpha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.5)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

### Plot Size

```python
plt.figure(figsize=(15,10))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

### Annotations

```python
df = pd.read_csv('batter.csv')
sample_df = df.sample(25)
sample_df
```

```python
x = [1,2,3,4]
y=[4,5,6,7]
plt.scatter(x,y)
plt.text(1,4,'point 1')
plt.text(2,5,'point 2')
plt.text(3,6,'point 3')
plt.text(4,7,'point 4',fontdict={'size':20,'color':'red'})
plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':20,'color':'green'})

plt.show()
```

### Horizontal and Vertical line

```python
plt.figure(figsize=(20,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':20,'color':'green'})

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,8))

ax[0].scatter(df['avg'],df['strike_rate'],color='g')
ax[1].scatter(df['avg'],df['runs'],color='r')

ax[0].set_title('avg vs strike_rate')
ax[0].set_ylabel('strike_rate')

ax[1].set_title('avg vs runs')
ax[1].set_xlabel('avg')
ax[1].set_ylabel('runs')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].scatter(df['avg'],df['strike_rate'],color='g')
ax[0,1].scatter(df['avg'],df['runs'],color='r')
ax[1,0].hist(df['avg'],color='r')
ax[1,1].hist(df['runs'],color='r')


plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['strike_rate'],color='g')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(df['avg'],df['runs'],color='g')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(df['avg'])

ax4 = fig.add_subplot(2,2,4)
ax4.hist(df['runs'])

plt.show()
```

### 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100 )
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.sin(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
z = 1/(1+np.exp(-xx)) + 1/(1+np.exp(-yy))
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

### Contour Plot
> 3D to 2D

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

### Heatmap

```python
df = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
df
```

```python
temp_df = df[(df['ballnumber'].isin([1,2,3,4,5,6])) & (df['batsman_run']==6)]
temp_df.head()
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(15,10))
plt.imshow(grid)
plt.title('Distribution of sixes with overs and ballNumber')
plt.xlabel('ballNumber')
plt.ylabel('OverNumber')
plt.yticks(df['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
plt.show()
```
