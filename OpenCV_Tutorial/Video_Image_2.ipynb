### saving video

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)

f = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("demo.mp4",f,40.0,(640,360))

while cap.isOpened():
    r, frame = cap.read()
    if r == True:
        frame = cv2.flip(frame,1)
        out.write(frame)
        cv2.imshow("vikas",frame)
        if cv2.waitKey(25) & 0xff == ord("p"):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

f = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("demo.mp4",f,40.0,(640,360),0)

while cap.isOpened():
    r, frame = cap.read()
    if r == True:
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frame = cv2.flip(frame,1)
        out.write(frame)
        cv2.imshow("vikas",frame)
        if cv2.waitKey(25) & 0xff == ord("p"):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread("images.jpeg")
img = cv2.resize(img,(500,400))

def wscube(a):
    pass

cv2.namedWindow("demo")
cv2.createTrackbar("lb","demo",0,255,wscube)
cv2.createTrackbar("lg","demo",0,255,wscube)
cv2.createTrackbar("lr","demo",0,255,wscube)

cv2.createTrackbar("ub","demo",0,255,wscube)
cv2.createTrackbar("ug","demo",0,255,wscube)
cv2.createTrackbar("ur","demo",0,255,wscube)

while True:
    hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

    Lb = cv2.getTrackbarPos("lb","demo")
    Lg = cv2.getTrackbarPos("lg","demo")
    Lr = cv2.getTrackbarPos("lg","demo")
    
    Ub = cv2.getTrackbarPos("ub","demo")
    Ug = cv2.getTrackbarPos("ug","demo")
    Ur = cv2.getTrackbarPos("ur","demo")
    
    lo = np.array([Lb,Lg,Lr])
    up = np.array([Ub,Ug,Ur])
    
    masks = cv2.inRange(hsv_img,lo,up)
    
    res = cv2.bitwise_and(img,img,mask=masks)
    
    cv2.imshow("res",res)
    cv2.imshow("mask",masks)
    cv2.imshow("hsv",hsv_img)
    cv2.imshow("wscube",img)
    if cv2.waitKey(1) & 0xff==ord("p"):
        break

cv2.destroyAllWindows()
```

### Histogram Equalization

### CLAHE (Contrast Limited Adaptive Histogram Equalizer)

```python
import cv2
```

```python
img = cv2.imread("histogram-equalize.jpg")

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img],[0],None,[255],[0,255])
eq = cv2.equalizeHist(gry)
hist = cv2.calcHist([eq],[0],None,[255],[0,255])

# plt.plot(eq,c = 'red')
# plt.plot(hist)
# plt.show()

cv2.imshow("vikas1",eq)
cv2.imshow("ikas",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("Screenshot 2025-10-01 155918.png")

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img],[0],None,[255],[0,255])
eq = cv2.equalizeHist(gry)
hist = cv2.calcHist([eq],[0],None,[255],[0,255])

cl = cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))
f = cl.apply(gry)

# plt.plot(eq,c = 'red')
# plt.plot(hist)
# plt.show()

cv2.imshow("wscube",f)
cv2.imshow("vikas1",eq)
cv2.imshow("ikas",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Mouse Event

```python
import cv2
import numpy as np
```

```python
def python(event,x,y,f,p):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),5,(0,0,255),4)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img,(x,y),(x+20,y+20),(255,0,0),4)
```

```python
cv2.namedWindow("wscube")
img = np.ones((600,600,3),np.uint8)*255
cv2.setMouseCallback("wscube",python)

while True:
    cv2.imshow("wscube",img)
    if cv2.waitKey(25) & 0xff==ord("p"):
        break

cv2.destroyAllWindows()
```

```python
cv2.imshow("vikas",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
