### Text Over a Image

```python
import cv2
```

```python
img_get = cv2.imread(r"D:\Project\project\Face_Recognition\ViaksKasaudhan\vikas1.jpeg")
img_get = cv2.resize(img_get,(400,400))


txt = cv2.putText(
    img = img_get,
    text = "Vikas Kasaudhan",
    org = (50,50),
    fontFace = cv2.FORMATTER_FMT_MATLAB,
    fontScale = 2,
    color = (255,0,0),
    thickness = 2,
    lineType = cv2.LINE_8,
    bottomLeftOrigin = False
)


txt = cv2.putText(
    img = img_get,
    text = "Vikas Kasaudhan",
    org = (50,50),
    fontFace = cv2.FORMATTER_FMT_MATLAB,
    fontScale = 2,
    color = (0,0,255),
    thickness = 2,
    lineType = cv2.LINE_8,
    bottomLeftOrigin = True
)

cv2.imshow('vikas',txt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Arithmetic Operation on Images

```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread("vikas1.jpg")
img2 = cv2.imread("vikas2.jpeg")

img1 = cv2.resize(img1,(400,400))
img2 = cv2.resize(img2,(400,400))

new_img = cv2.addWeighted(img1,1,img2,0.2,5)
cv2.imshow("Vikas",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("vikas1.jpg")
img2 = cv2.imread("vikas2.jpeg")

img1 = cv2.resize(img1,(400,400))
img2 = cv2.resize(img2,(400,400))

new_img = cv2.subtract(img1,img2)
cv2.imshow("Vikas",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Bitwise Operation on images

```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread("images1.png")
img2 = cv2.imread("images2.png")

img1 = cv2.resize(img1,(400,400))
img2 = cv2.resize(img2,(400,400))

new_img = cv2.bitwise_and(img1,img2)
# new_img = cv2.bitwise_or(img1,img2)
#new_img = cv2.bitwise_not(img1)
# new_img = cv2.bitwise_xor(img1,img2)

h = np.hstack((img1,img2,new_img))

cv2.imshow("Vikas",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("images1.png")

img1 = cv2.resize(img1,(400,400))

new_img = cv2.bitwise_not(img1)

h = np.hstack((img1,new_img))

cv2.imshow("Vikas",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("vikas1.jpg")
img2 = cv2.imread("vikas2.jpeg")

img1 = cv2.resize(img1,(400,400))
img2 = cv2.resize(img2,(400,400))

#new_img = cv2.bitwise_and(img1,img2)
#new_img = cv2.bitwise_or(img1,img2)
new_img = cv2.bitwise_xor(img1,img2)

h = np.hstack((img1,img2,new_img))

cv2.imshow("Vikas",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Edge Detection

```python
img = cv2.imread("vikas1.jpg")
print(img.shape)

img = cv2.resize(img,(400,400))

new_img = cv2.Canny(img,400,200,apertureSize=5,L2gradient=True)
print(new_img.shape)


cv2.imshow("Vikas",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Scaling and Rotation of images

```python
import cv2
```

```python
ori_img = cv2.imread("vikas1.jpg")
new_img = cv2.resize(ori_img,(500,500))
w,h = new_img.shape[0],new_img.shape[1]

m = cv2.getRotationMatrix2D((w/2,h/2),180,1)
fin_img = cv2.warpAffine(new_img,m,(h,w))

h = np.hstack((new_img,fin_img))
cv2.imshow("vikas",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image Blurring 

```python
import cv2
import numpy as np
```

```python
ori_img = cv2.imread("vikas1.jpg")
new_img = cv2.resize(ori_img,(300,300))

g = cv2.GaussianBlur(new_img,(5,5),0)
m = cv2.medianBlur(new_img,5)
b = cv2.bilateralFilter(new_img,10,100,100)

h = np.hstack((new_img,g,m,b))
cv2.imshow("vikas",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Imwrite Function (saving images)

```python
ori_img1 = cv2.imread("vikas1.jpg")
ori_img2 = cv2.imread("vikas2.jpeg")
ori_img3 = cv2.imread("vikas3.jpg")
ori_img4 = cv2.imread("vikas8.jpg")

new_img1 = cv2.resize(ori_img1,(300,300))
new_img2 = cv2.resize(ori_img2,(300,300))
new_img3 = cv2.resize(ori_img3,(300,300))
new_img4 = cv2.resize(ori_img4,(300,300))

h1 = np.hstack((new_img1,new_img2))
h2 = np.hstack((new_img3,new_img4))
v = np.vstack((h1,h2))

cv2.imwrite("Kasaudhan.png",v)

cv2.imshow("vikas",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
