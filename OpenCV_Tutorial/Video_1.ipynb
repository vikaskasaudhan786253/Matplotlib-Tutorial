### Play a video using OpenCV

```python
import cv2
```

```python
cap = cv2.VideoCapture(r"C:\Users\vikas\Videos\Captures\2.mp4")

while cap.isOpened():
    r, frame = cap.read()
    if r==True:
        frame = cv2.resize(frame,(400,500))
        cv2.imshow("vikas",frame)
        if cv2.waitKey(25) & 0xff == ord("p"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindow()
```

### Morphological operations using OpenCV part -  1

####  Erosion and Dilation

```python
import cv2
import numpy as np
```

```python
img = cv2.imread("J.png")

m = np.ones((15,15),np.int8)

er = cv2.erode(img,m,iterations=1)
di = cv2.dilate(img,m,iterations=1)

cv2.imshow("vikas",img)
cv2.imshow("vikas1",er)
cv2.imshow("vikas2",di)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Closing and Opening and gradient

```python
img = cv2.imread("opening.png")

k = np.ones((10,10),np.int8)

op = cv2.morphologyEx(img,cv2.MORPH_OPEN,k,iterations=1)


cv2.imshow("vikas",img)
cv2.imshow("vikas2",op)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("closing.png")

k = np.ones((10,10),np.int8)

cl = cv2.morphologyEx(img,cv2.MORPH_CLOSE,k,iterations=1)


cv2.imshow("vikas",img)
cv2.imshow("vikas2",cl)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("J.png")

k = np.ones((10,10),np.int8)

gr = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,k,iterations=1)

cv2.imshow("vikas",img)
cv2.imshow("vikas1",gr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Background subtraction

```python
import cv2
```

```python
org_v = cv2.VideoCapture("loop.mp4")

sub_m = cv2.createBackgroundSubtractorMOG2()

while True:
    r, frame = org_v.read()
    if r == True:
        frame = cv2.resize(frame,(700,500))
        sub_v = sub_m.apply(frame)
        cv2.imshow("vikas2",sub_v)
        cv2.imshow("vikas",frame)
        if cv2.waitKey(25) & 0xff==ord("p"):
            break
    else:
        break

org_v.release()
cv2.destroyAllWindows()
```
