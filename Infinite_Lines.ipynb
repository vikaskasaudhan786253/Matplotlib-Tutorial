```python
import numpy as np
import matplotlib.pyplot as plt
```

* `axvline` -> Infinite verticle line
* `axhline` -> infinite horizontal line
* `axline` -> infinite straight lines in in arbitrary direction

```python
t = np.linspace(-10,10,50)
sig = 1/(1+np.exp(-t))

fig, ax = plt.subplots()

ax.axhline(y=0,color='black',linestyle='--')
ax.axhline(y=1,color='black',linestyle='--')
ax.axhline(y=0.5,color='black',linestyle=':')
ax.axvline(x=0,color='black')
ax.axline((0,0.5),slope=0.25,color='black',linestyle='--')

ax.plot(t,sig,linewidth=2,color='g',label=r"$\sigma(t) = \frac{1}{1+e^{-t}}$")
ax.set(xlim=(-10,10),xlabel='t')
ax.legend(fontsize=15)

plt.show()
```

`axline` can also be used with a transform parameter, which applies to the point, but not to the slope. This can be useful for drawing diagonal grid lines with a fixed slope, which stay in place when the plot limits are moved.

```python
fig , ax = plt.subplots()

for pos in np.linspace(-2,1,10):
    ax.axline((pos,0), slope=0.5, color='green')

plt.show()
```

```python
fig , ax = plt.subplots()

for pos in np.linspace(-2,1,10):
    ax.axline((pos,0), slope=0.5, color='green')

ax.set(xlim=(0,1),ylim=(0,1))
plt.show()
```

```python
fig , ax = plt.subplots()

for pos in np.linspace(-2,1,10):
    ax.axline((pos,0), slope=0.5, color='green',transform=ax.transAxes)

ax.set(xlim=(0,1),ylim=(0,1))
plt.show()
```
